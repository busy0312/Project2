```python
#Import dependencies
import pandas as pd
from datetime import timedelta
from sqlalchemy import create_engine
from config import SQL_key
```

```python
# Read csv file
path_intake="raw_data/Austin_Animal_Center_Intakes.csv"
path_outcome="raw_data/Austin_Animal_Center_Outcomes.csv"
```

```python
animal_outcome=pd.read_csv(path_outcome)
animal_outcome.head()
```

```python
# drop columns
animal_outcome_df=animal_outcome.drop(columns=['Outcome Subtype','MonthYear'])
animal_outcome_df.head()
```

```python
animal_outcome_df[animal_outcome_df['Animal ID']=='A689724']
```

```python
# load file
animal_intake=pd.read_csv(path_intake)
animal_intake.head()
```

```python
#drop columns
animal_intake_df=animal_intake.drop(columns=['MonthYear','Breed','Color','Name','Intake Condition','Animal Type',"Sex upon Intake"])
animal_intake_df.head()
```

```python
animal_intake_df[animal_intake_df['Animal ID']=='A689724']
```

```python
# merge two csv
combined=pd.merge(animal_outcome_df,animal_intake_df,on="Animal ID",how="left")
combined.head()
```

```python
#rename columns and sort by date
combined_df=combined.rename(columns={
    "DateTime_x": "DateTime_Outcome", "DateTime_y": "DateTime_Intake","Animal ID":"Animal_ID",
    "Date of Birth":"Date_of_Birth","Outcome Type":"Outcome_Type","Animal Type":"Animal_Type",
    "Sex upon Outcome":"Sex_upon_Outcome","Age upon Outcome":"Age_upon_Outcome",
    "Intake Type":"Intake_Type","Age upon Intake":"Age_upon_Intake","Found Location":"Found_Location"

})
combined_df=combined_df.sort_values(by='Animal_ID',ascending=False)
combined_df.head()
```

```python
# duplicated columns
combined_df=combined_df.drop_duplicates(subset='Animal_ID', keep='first')
# only keep Dogs and Cats
combined_df=combined_df[(combined_df['Animal_Type']=='Dog')|(combined_df['Animal_Type']=='Cat')]
```

```python
# Combined mix 
combined_df["Breed"]=combined_df["Breed"].str.replace("/",' Mix /')
combined_df["Breed"]=combined_df["Breed"].str.partition("/")
combined_df.head()
```

```python
# function that convert dateunit
def date_func(d):
    if type(d) == str:
        num = d.split(" ")[0]
        days = 0
        date_type = d.split(" ")[1]
        if date_type == "weeks":
            days = 7 * int(num)
        elif date_type == "week":
            days = 7 * int(num)
        elif date_type == "month":
            days = 30 * int(num)
        elif date_type == "months":
            days = 30 * int(num)
        elif date_type == "year":
            days = 365 * int(num)    
        elif date_type == "years":
            days = 365 * int(num)
        elif date_type == "days":
            days = 1 * int(num)
        elif date_type == "day":
            days = 1 * int(num)
        return days
```

```python
combined_df['Age_upon_Outcome_days'] = combined_df['Age_upon_Outcome'].map(date_func)
combined_df['Age_upon_Intake_days'] = combined_df['Age_upon_Intake'].map(date_func)
```

```python
# get rid of decimal 
combined_df['Age_upon_Outcome_days']=combined_df['Age_upon_Outcome_days'].map('{:.0f}'.format)
combined_df['Age_upon_Intake_days']=combined_df['Age_upon_Intake_days'].map('{:.0f}'.format)
combined_df.head()
```

```python
combined_df['Age_upon_Outcome'].str.contains("month|weeks")
```

```python
# Push transformed CSV file to folder
combined_df.to_csv("Animals.csv", index=False, header=True)
```
